## Walking Speed Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
data = pd.read_csv("ms_data_with_insurance.csv")
```

```python
# Scatter plot of age vs walking speed with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='walking_speed', data=data, scatter=True, line_kws={'color': 'red'})
plt.title('Age vs Walking Speed')
plt.xlabel('Age')
plt.ylabel('Walking Speed')
plt.show()
```

```python
# Box plots by education level
plt.figure(figsize=(10, 6))
sns.boxplot(x='education_level', y='walking_speed', data=data)
plt.title('Walking Speed by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Walking Speed')
plt.show()
```

```python
# Line plot showing education age interaction
plt.figure(figsize=(15, 6))
sns.lineplot(x='age', y='walking_speed', hue='education_level', data=data)
plt.title('Age and Education Level on Walking Speed')
plt.xlabel('Age')
plt.ylabel('Walking Speed')
plt.show()
```

## Cost analysis

```python
# Bar plot of mean costs by insurance type
plt.figure(figsize=(10, 6))
sns.barplot(x='insurance_type', y='visit_cost', data=data, errorbar='sd')
plt.title('Mean Costs by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Mean Cost')
plt.show()
```

```python
# Box plots showing cost distributions
plt.figure(figsize=(10, 6))
sns.boxplot(x='insurance_type', y='visit_cost', data=data)
plt.title('Cost Distributions by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Visit Cost')
plt.show()
```

## Combined Visualizations

```python
# Pair plot of key variables
sns.pairplot(data, vars=['age', 'walking_speed', 'visit_cost'], hue='education_level')
plt.suptitle('Pair Plot of Key Variables', y=1.02)
plt.show()
```

```python
# Faceted plots by education/insurance
g = sns.FacetGrid(data, col='education_level', row='insurance_type', margin_titles=True)
g.map_dataframe(sns.scatterplot, x='age', y='walking_speed')
g.set_axis_labels('Age', 'Walking Speed')
g.set_titles(col_template="{col_name}", row_template="{row_name}")
g.figure.suptitle('Walking Speed by Education, Insurance, and Age')
g.figure.subplots_adjust(top=0.9)
plt.show()
```

```python
# Time trends
plt.figure(figsize=(14, 6))
sns.lineplot(x='visit_date', y='visit_cost', hue='insurance_type', data=data)
plt.title('Time Trend of Cost by Insurance Type')
plt.xlabel('Date')
plt.ylabel('Cost')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))  
plt.gcf().autofmt_xdate()  
plt.show()
```
